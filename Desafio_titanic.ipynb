```python
# Limpando os dados
#Achei melhor excluir os dados Na
dados = dados.drop_duplicates()
dados = dados.drop(columns=['Cabin','Name','PassengerId','Ticket'])
dados = dados.dropna()
dados.isnull().sum()
```

```python
#Transformando dados coluna sex (male = 1, female = 0), coluna Embarked (c = 1, q =2 e s =3) e Age para int
dados['Sex'] = dados['Sex'].map(lambda x:1 if x=='male' else 0)
dados['Embarked'] = dados['Embarked'].map(lambda x:1 if x=='C' else (2 if x=='Q' else 3))
dados['Age'] = dados['Age'].astype(int)
```

```python
#Normalizei só para fazer o feature selection
# Import dos módulos
from pandas import read_csv
from sklearn.preprocessing import MinMaxScaler

df = dados
array = df.values

# Separando o array em componentes de input e output
X = array[:,1:8]
Y = array[:,0]

# Gerando a nova escala
scaler = MinMaxScaler(feature_range = (0, 1))
rescaledX = scaler.fit_transform(X)

# Sumarizando os dados transformados
print(rescaledX[0:5,:])
```

```python
#PCA (Os dados precisam estar normalizados (em escala))
# Feature Extraction

# Import dos módulos
from pandas import read_csv
from sklearn.decomposition import PCA

# Separando o array em componentes de input e output
X = rescaledX

# Seleção de atributos
pca = PCA(n_components = 9)
fit = pca.fit(X)

# Sumarizando os componentes
print(sum(fit.explained_variance_ratio_))
print(fit.explained_variance_ratio_)
X = fit.transform(X)
```

```python
# Feature selection
# Extração de Variáveis com Testes Estatísticos Univariados (Teste qui-quadrado)

# Import dos módulos
from pandas import read_csv
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = rescaledX

# Extração de Variáveis
test = SelectKBest(score_func = chi2, k = 4)
fit = test.fit(X, Y)

# Sumarizando o score
print(fit.scores_)
```

```python
# Feature selection
# Importância do Atributo com o Extra Trees Classifier

# Import dos Módulos
from pandas import read_csv
from sklearn.ensemble import ExtraTreesClassifier

X = rescaledX


# Criação do Modelo - Feature Selection
modelo = ExtraTreesClassifier()
modelo.fit(X, Y)

# Print dos Resultados
#print(df.columns[0:20])
print(modelo.feature_importances_)
```

```python
# Padronizando os dados (0 para a média, 1 para o desvio padrão)

# Import dos módulos
from pandas import read_csv
from sklearn.preprocessing import StandardScaler

df = dados
array = df.values

# Separando o array em componentes de input e output
X = array[:,1:8]
Y = array[:,0]

#Após features selection, escolher as features que serão utilizadas (fiz isso antes de normalizar)
X = X[:,[0,1,2,5]]


# Gerando o novo padrão
scaler = StandardScaler().fit(X)
standardX = scaler.transform(X)

# Sumarizando os dados transformados
print(standardX[1:20,:])
```

```python
import plotly.express as px
df = standardX[1:1000]
fig = px.histogram(df)
fig.show()
```

```python
# Avaliação usando dados de treino e de teste

# Import dos módulos
from sklearn import model_selection
from sklearn.naive_bayes import GaussianNB


# Separando o array em componentes de input e output
X = standardX

# Definindo o tamanho das amostras
teste_size = 0.33

# Garante que os resultados podem ser reproduzidos
# Isso é importante para comparar a acurácia com outros algoritmos de Machine Learning.
seed = 7

# Criando os conjuntos de dados de treino e de teste
X_treino, X_teste, Y_treino, Y_teste = model_selection.train_test_split(X, Y, 
                                                                         test_size = teste_size, 
                                                                         random_state = seed)
# Definindo os valores para o número de folds
num_folds = 10
num_instances = len(X_treino)


# Separando os dados em folds
kfold = model_selection.KFold(num_folds, shuffle=True, random_state = seed)

# Criando o modelo
modelo = GaussianNB()
resultado = model_selection.cross_val_score(modelo, X, Y, cv = kfold, scoring='f1')

# Print do resultado
print(resultado.mean())
```

```python
# Import dos módulos
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report

# Criando o modelo
model = GaussianNB()
model.fit(X_treino, Y_treino)

#Validação Cruzada
# Definindo os valores para o número de folds
num_folds = 10
num_instances = len(X_treino)

# Separando os dados em folds
kfold = model_selection.KFold(num_folds, shuffle=True)
resultado = model_selection.cross_val_score(model, X, Y, cv = kfold, scoring='f1')
print(resultado.mean())

# Fazendo as previsões e construindo o relatório
previsoes = model.predict(X_teste)
report = classification_report(Y_teste, previsoes)

# Imprimindo o relatório
print(report)
```

```python
# Import dos módulos
from pandas import read_csv
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Criando o modelo
modelo = XGBClassifier()
modelo.fit(X_treino, Y_treino)

# Pront do modelo
print(modelo)

# Fazendo previsões
y_pred = modelo.predict(X_teste)
previsoes = [round(value) for value in y_pred]


#Validação Cruzada
# Definindo os valores para o número de folds
num_folds = 10
num_instances = len(X_treino)

# Separando os dados em folds
kfold = model_selection.KFold(num_folds, shuffle=True)
resultado = model_selection.cross_val_score(modelo, X, Y, cv = kfold, scoring='f1')
print(resultado.mean())

# Avaliando as previsões
accuracy = accuracy_score(Y_teste, previsoes)
print("Acurácia: %.2f%%" % (accuracy * 100.0))
```
